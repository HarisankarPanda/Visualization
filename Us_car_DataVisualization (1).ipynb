```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("/content/us_car_data.csv")
df.info()
```

```python
plt.figure(figsize = (15,15))
sns.countplot(x='Model',data=df)
```

```python
plt.figure(figsize = (12,8))
sns.countplot(x='Model Year',data = df)
```

```python
fig, ax = plt.subplots(figsize=(15, 20))
sns.set_theme(style="white")
plt.pie(x='Model',y='Model Year',data=df,ax=ax)
plt.title('MODEL VS MODEL YEAR',color= 'lightblue',size=15)
```

```python
sns.set_theme(style="dark")
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="Longitude", y="latitude",

                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=df, ax=ax)

plt.title('LOCATION',color='red',size=18)
```

```python
sns.set_theme(style="dark")
palette = sns.color_palette("rocket_r")
sns.lineplot( x="Electric Range", y="Model",data=df,palette=palette)
```
